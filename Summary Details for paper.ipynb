```python
def read_in_data(diagnoses_csv,std,MeSH,sno_all,hierarchy):
    
    diagnoses = pd.read_csv(diagnoses_csv)
    fstd = open(std)
    fmsh = open(MeSH)
    fall = open(sno_all)
    fhrc = open(hierarchy)
    
    
    std_diag_dict = json.load(fstd)
    diag_MeSH_dict = json.load(fmsh)
    all_snomed_diag = json.load(fall)
    hierarchy_diag = json.load(fhrc)
    
    
    alt_hierarchy = {}
    for ICD in hierarchy_diag:
        snomeds = []
        for snomed_type in hierarchy_diag[ICD]:
            snomeds = snomeds + hierarchy_diag[ICD][snomed_type]
        alt_hierarchy[ICD] = snomeds
    
    return diagnoses,std_diag_dict,diag_MeSH_dict,all_snomed_diag,alt_hierarchy
```
