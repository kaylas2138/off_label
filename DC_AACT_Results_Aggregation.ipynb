```python
def read_in_data(diagnoses_csv,std,MeSH,sno_all,hierarchy):
    
    diagnoses = pd.read_csv(diagnoses_csv)
    fstd = open(std)
    fmsh = open(MeSH)
    fall = open(sno_all)
    fhrc = open(hierarchy)
    
    
    std_diag_dict = json.load(fstd)
    diag_MeSH_dict = json.load(fmsh)
    all_snomed_diag = json.load(fall)
    hierarchy_diag = json.load(fhrc)
    
    
    alt_hierarchy = {}
    for ICD in hierarchy_diag:
        snomeds = []
        for snomed_type in hierarchy_diag[ICD]:
            snomeds = snomeds + hierarchy_diag[ICD][snomed_type]
        alt_hierarchy[ICD] = snomeds
    
    return diagnoses,std_diag_dict,diag_MeSH_dict,all_snomed_diag,alt_hierarchy
```

```python
DC_indications = pd.read_csv('Gabapentin_indications_drugcentral.csv', sep = ',', header =0)
if True in DC_indications.columns.str.match("Unnamed"):
    DC_indications = DC_indications.drop('Unnamed: 0',axis=1)

# Convert all SNOMED ID's from DrugCentral to lists for each classification of indication, off-label use, contraindication
DC_SNOMED_ind = (DC_indications.loc[DC_indications['relationship_name']=='indication'])['snomed_concept_id'].values.tolist()
DC_SNOMED_off = (DC_indications.loc[DC_indications['relationship_name']=='off-label use'])['snomed_concept_id'].values.tolist()
DC_SNOMED_contr = (DC_indications.loc[DC_indications['relationship_name']=='contraindication'])['snomed_concept_id'].values.tolist()
```

```python
# TO DO: for now just prints, create dictionary output, or whatever makes the most sense 
# TO DO: count matches

# Input dictionary of ICD codes and SNOMED codes
# Output which SNOMED codes map to DrugCentral Indication data
def get_DC_matches(std_diag_dict):
    
    ind = []
    off = []
    contr = []
    unmapped = []
    
    for i in std_diag_dict.values(): # iterate through ICD codes
        for j in i: # for all associated SNOMED codes 
            if j in DC_SNOMED_ind: # an indication
                ind.append(j)
            elif j in DC_SNOMED_off: # off label use
                off.append(j)
            elif j in DC_SNOMED_contr: # contraindication
                contr.append(j)
            else:
                unmapped.append(j)
                
    results_dict = {}
    results_dict['Indications'] = ind
    results_dict['Off-Label'] = off
    results_dict['Contraindications'] = contr
    results_dict['Unmapped'] = unmapped
    
    
    
    return results_dict
```

```python
def DrugCentral_Stats(diagnoses,DC_indications,std_diag,all_diag,hierarchy_diag):
    
    std_DC = get_DC_matches(std_diag)
    NonS_DC = get_DC_matches(all_diag)
    hierarchy_DC = get_DC_matches(hierarchy_diag)
    
    patient_dict = {}
    
    patient_dict['patientID'] = diagnoses['subject_id'][0]
    patient_dict['visitID'] = diagnoses['hadm_id'][0]
    patient_dict['ICD_Count'] = len(std_diag)

    Std_DC_Ind = len(std_DC['Indications'])
    Std_DC_Off = len(std_DC['Off-Label'])
    Std_DC_Contr = len(std_DC['Contraindications'])
    Std_DC_Unmapped = len(std_DC['Unmapped'])
    Std_Total_DC = Std_DC_Ind + Std_DC_Off + Std_DC_Contr

    patient_dict['Standard_DC_Ind'] = Std_DC_Ind
    patient_dict['Standard_DC_Off'] = Std_DC_Off
    patient_dict['Standard_DC_Contr'] = Std_DC_Contr
    patient_dict['Standard_DC_Unmapped'] = Std_DC_Unmapped
    patient_dict['Std_Total_DC'] = Std_Total_DC

    NonS_DC_Ind = len(NonS_DC['Indications'])
    NonS_DC_Off = len(NonS_DC['Off-Label'])
    NonS_DC_Contr = len(NonS_DC['Contraindications'])
    NonS_DC_Unmapped = len(NonS_DC['Unmapped'])
    NonS_Total_DC = NonS_DC_Ind + NonS_DC_Off + NonS_DC_Contr

    patient_dict['NonS_DC_Ind'] = NonS_DC_Ind
    patient_dict['NonS_DC_Off'] = NonS_DC_Off
    patient_dict['NonS_DC_Contr'] = NonS_DC_Contr
    patient_dict['NonS_DC_Unmapped'] = NonS_DC_Unmapped
    patient_dict['NonS_Total_DC'] = NonS_Total_DC

    Hierarchy_DC_Ind = len(hierarchy_DC['Indications'])
    Hierarchy_DC_Off = len(hierarchy_DC['Off-Label'])
    Hierarchy_DC_Contr = len(hierarchy_DC['Contraindications'])
    Hierarchy_DC_Unmapped = len(hierarchy_DC['Unmapped'])
    Hierarchy_Total_DC = Hierarchy_DC_Ind + Hierarchy_DC_Off + Hierarchy_DC_Contr

    patient_dict['Hier_DC_Ind'] = Hierarchy_DC_Ind
    patient_dict['Hier_DC_Off'] = Hierarchy_DC_Off
    patient_dict['Hier_DC_Contr'] = Hierarchy_DC_Contr
    patient_dict['Hier_DC_Unmapped'] = Hierarchy_DC_Unmapped
    patient_dict['Hierarchy_Total_DC'] = Hierarchy_Total_DC
    
    patient_dict['Indications'] = list(set(std_DC['Indications'] + NonS_DC['Indications'] + hierarchy_DC['Indications']))
    patient_dict['Off-Label'] = list(set(std_DC['Off-Label'] + NonS_DC['Off-Label'] + hierarchy_DC['Off-Label']))
    

    return patient_dict
```

```python
def get_AACT_matches(*args):
    if len(args) == 1:
        
        diag_MeSH_dict = args[0]
        
        listAll = []

        for ICD in diag_MeSH_dict:
            concept = diag_MeSH_dict[ICD]['MeSH_concept'][0]


            if (gb_AACT.loc[gb_AACT['cond_mesh'] == concept]).empty == False:
                #display(gb_AACT.loc[gb_AACT['cond_mesh'] == concept])
                #listTemp = gb_AACT.loc[gb_AACT['cond_mesh'] == concept].values.tolist()
                listTemp = gb_AACT.loc[gb_AACT['cond_mesh'] == concept].values.tolist()
                listAll = listAll + listTemp

        AACT_df = pd.DataFrame(listAll, columns = ['nct_id','drug_mesh','cond_mesh','mesh_type'])

    elif len(args) == 2:
        
        diag_MeSH_dict = args[0]
        diagnoses = args[1]

        listAll = []

        for ICD in diag_MeSH_dict:
            concept = diag_MeSH_dict[ICD]['MeSH_concept'][0]


            if (gb_AACT.loc[gb_AACT['cond_mesh'] == concept]).empty == False:
                #display(gb_AACT.loc[gb_AACT['cond_mesh'] == concept])
                #listTemp = gb_AACT.loc[gb_AACT['cond_mesh'] == concept].values.tolist()
                listTemp = gb_AACT.loc[gb_AACT['cond_mesh'] == concept].values.tolist()
                listAll = listAll + listTemp

        AACT_df = pd.DataFrame(listAll, columns = ['nct_id','drug_mesh','cond_mesh','mesh_type'])
        AACT_df['patient_id'] = diagnoses['subject_id'][0]

    return AACT_df
```
